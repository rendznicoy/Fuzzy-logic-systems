```python
from fuzzylogic.classes import Domain, Set, Rule
from matplotlib import pyplot
from fuzzylogic.functions import R, S, alpha
from fuzzylogic.functions import (trapezoid, triangular)
```

```python
time_of_day = Domain("Time of Day", 0, 23, res = 0.1)
time_of_day.midnight = S(0, 6)
time_of_day.midnight.plot()
time_of_day.morning = triangular(2, 12)
time_of_day.morning.plot()
time_of_day.noon = triangular(10, 14)
time_of_day.noon.plot()
time_of_day.afternoon = triangular(13, 17)
time_of_day.afternoon.plot()
time_of_day.evening = R(16, 23)
time_of_day.evening.plot()

pyplot.legend(["MIDNIGHT", "MORNING", "NOON", "AFTERNOON", "EVENING"])
pyplot.title("Time of Day", fontweight = 'bold', fontsize = 15)
pyplot.show()
```

```python
ambient_level = Domain("Ambient Level", 0, 100, res = 0.1)
ambient_level.verydark = S(0, 30)
ambient_level.verydark.plot()
ambient_level.dark = triangular(20, 50)
ambient_level.dark.plot()
ambient_level.dim = triangular(40, 70)
ambient_level.dim.plot()
ambient_level.well_lit = triangular(60, 90)
ambient_level.well_lit.plot()
ambient_level.too_lit = R(90, 100)
ambient_level.too_lit.plot()

pyplot.legend(["VERY DARK", "DARK", "DIM", "WELL-LIT", "TOO-LIT"])
pyplot.title("Ambient Light Level", fontweight = 'bold', fontsize = 15)
pyplot.show()
```

```python
brightness_level = Domain("Brightness Level", 0, 100, res = 0.1)
brightness_level.verylow = S(0, 20)
brightness_level.verylow.plot()
brightness_level.low = triangular(10, 50)
brightness_level.low.plot()
brightness_level.medium = triangular(40, 80)
brightness_level.medium.plot()
brightness_level.high = triangular(70, 90)
brightness_level.high.plot()
brightness_level.veryhigh = R(80, 100)
brightness_level.veryhigh.plot()

pyplot.legend(["Very Low", "Low", "Medium", "High", "Very High"])
pyplot.title("Brightness Level", fontweight ='bold', fontsize = 15)
pyplot.show()
```
